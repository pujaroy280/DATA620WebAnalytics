```python
%matplotlib inline

import networkx as nx
import matplotlib.pyplot as plt

# Create an empty graph
G = nx.Graph()

# Add connections (edges) between people (nodes)
edges = [
    ('andre', 'beverly'), ('andre', 'carol'), ('andre', 'diane'),
    ('beverly', 'andre'), ('beverly', 'diane'), ('beverly', 'ed'),
    ('carol', 'andre'), ('carol', 'diane'), ('carol', 'fernando'),
    ('fernando', 'garth'), ('fernando', 'diane'), ('fernando', 'carol'),
    ('garth', 'ed'), ('garth', 'diane'), ('garth', 'fernando'),
    ('garth', 'heather'), ('ed', 'beverly'), ('ed', 'diane'),
    ('ed', 'garth'), ('heather', 'fernando'), ('heather', 'garth'),
    ('heather', 'ike'), ('ike', 'heather'), ('ike', 'jane'),
    ('jane', 'ike')
]
G.add_edges_from(edges)

# Set the positions of each person (node)
pos = {
    'andre': (0, 1), 'beverly': (0, -1), 'carol': (1, 2),
    'diane': (1, 0), 'ed': (1, -2), 'ike': (4, 0),
    'garth': (2, -1), 'jane': (5, 0), 'fernando': (2, 1),
    'heather': (3, 0)
}

# Draw the people (nodes)
nx.draw_networkx_nodes(G, pos, node_size=800, node_color='r')

# Draw the connections (edges) between people
nx.draw_networkx_edges(G, pos, alpha=0.5, width=6)

# Draw the names (labels) of the people
nx.draw_networkx_labels(G, pos, font_size=15, font_family='sans-serif')

# Remove the axes and display the graph
plt.axis('off')
plt.show()
```
